```python
import pandas as pd
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

### Données

```python
df=pd.read_csv('data.csv',sep=';')
```

```python
df.head()
```

Les données 'voteshare' sont en format texte 

```python
df['voteshare']=df['voteshare'].apply(lambda x : np.float(x.replace(',', '.')))
df['ab']=df['ab'].apply(lambda x : np.float(x.replace(',', '.')))
```

```python
df['score_beaute']=df[['im24', 'if39','nif24', 'nim40']].mean(axis=1)
df['score_beaute_int']=df['score_beaute'].apply(lambda x : np.int(x))
```

```python
df['note_i']=df[['im24', 'if39']].mean(axis=1)
df['note_ni']=df[['nif24','nim40']].mean(axis=1)

df['note_i_int']=df[['im24', 'if39']].mean(axis=1).apply(lambda x : np.int(x))
df['note_ni_int']=df[['nif24','nim40']].mean(axis=1).apply(lambda x : np.int(x))
```

```python
df['note_femme']=df[[ 'if39','nif24']].mean(axis=1)
df['note_homme']=df[['im24','nim40']].mean(axis=1)
```

```python
df['score_beaute_ab']=df['ab']*df['note_i']+(1-df['ab'])*df['note_ni']
df['score_beaute_ab_int']=df['score_beaute_ab'].apply(lambda x : np.int(x))
```

```python
df['skin_g']=pd.qcut(df['skin'],6,labels=False,duplicates='drop')
```

```python
df.columns
```

## Densités et Corrplot

```python
p_corr=df.drop(['score_beaute_int','note_i','note_ni','note_i_int','note_ni_int','skin_g','score_beaute_ab_int','im24', 'if39',
       'nif24', 'nim40','score_beaute_ab'],axis=1).corr()
```

### Evolutions

```python
liste=['skin','skin_g']
fig, axs = plt.subplots(nrows=1, ncols=2)
axs = axs.flatten()
fig.subplots_adjust(hspace=1, wspace=0.4)
fig.set_size_inches(18.5, 5)


for i in range(len(liste)):

    df_total_mean=df[['voteshare']+[liste[i]]].groupby(liste[i]).mean()
    df_total_var=df[['voteshare']+[liste[i]]].groupby(liste[i]).var()
    df_total_count=df[['voteshare']+[liste[i]]].groupby(liste[i]).count()

    ub=df_total_mean+(np.sqrt(df_total_var)*1.96)/(np.sqrt(df_total_count))
    lb=df_total_mean-(np.sqrt(df_total_var)*1.96)/(np.sqrt(df_total_count))

    
    axs[i].set_xlabel(liste[i])
    axs[i].set_ylabel('voteshare')
    axs[i].plot(df_total_mean,color='darkmagenta')
    axs[i].fill_between(df_total_mean.index, ub.iloc[:,0], lb.iloc[:,0],alpha=.35,color='orange')
```
